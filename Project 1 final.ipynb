```python
import pandas
```

```python
Data = pandas.read_csv('house_size_price_data.csv')
```

```python
Data
```

```python
X = Data[['House_Size_Sqft']]
Y = Data['House_Price']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,Y,'x')
plt.xlabel('Size')
plt.ylabel('Price')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
asad = LinearRegression()

```

```python
asad.fit(X,Y)
```

```python
asad.predict([[100]])
```

```python
import numpy as np
```

```python
x = np.linspace(0,5000,50000)
```

```python
y = asad.predict(x.reshape(len(x),1))
```

```python
plt.plot(X,Y,'x')
plt.plot(x,y)
plt.title('Asad prediction model for house')
plt.xlabel('Size')
plt.ylabel('Price')
```
